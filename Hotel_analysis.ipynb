# Time series analysis

```python

# Convert 'Date' column to datetime format
data['Date'] = pd.to_datetime(data['Date'], format='%d/%m/%Y')

# Set 'Date' as the index
data.set_index('Date', inplace=True)

# Select only numeric columns for the time series analysis
numeric_columns = ['Marketing_Spend', 'Revenue', 'Room_Revenue', 'Occupancy_Rate', 'ADR', 
                   'Occupancy_Managed_Guests', 'RevPAR_All', 'Occupancy_All', 
                   'Room_Revenue_All', 'Total_Revenue', 'Operating_Expenses', 
                   'Fixed_Costs', 'Variable_Costs', 'Total_Costs', 'Profit']

# Resample data to monthly frequency and calculate the mean for numeric columns
monthly_data = data[numeric_columns].resample('M').mean()

# Plotting monthly Revenue and Occupancy Rate
fig, ax1 = plt.subplots(figsize=(14, 7))

ax1.set_title('Monthly Revenue and Occupancy Rate')
ax1.set_xlabel('Date')

ax1.plot(monthly_data.index, monthly_data['Revenue'], color='tab:blue', label='Revenue')
ax1.set_ylabel('Revenue', color='tab:blue')
ax1.tick_params(axis='y', labelcolor='tab:blue')

ax2 = ax1.twinx()
ax2.plot(monthly_data.index, monthly_data['Occupancy_Rate'], color='tab:orange', label='Occupancy Rate')
ax2.set_ylabel('Occupancy Rate', color='tab:orange')
ax2.tick_params(axis='y', labelcolor='tab:orange')

fig.tight_layout()
plt.legend(loc='upper left')
plt.show()
```

```python
# Plotting multiple time series

# Plotting Marketing Spend and Revenue
fig, ax1 = plt.subplots(figsize=(14, 7))

ax1.set_title('Monthly Marketing Spend and Revenue')
ax1.set_xlabel('Date')

ax1.plot(monthly_data.index, monthly_data['Marketing_Spend'], color='tab:blue', label='Marketing Spend')
ax1.set_ylabel('Marketing Spend', color='tab:blue')
ax1.tick_params(axis='y', labelcolor='tab:blue')

ax2 = ax1.twinx()
ax2.plot(monthly_data.index, monthly_data['Revenue'], color='tab:orange', label='Revenue')
ax2.set_ylabel('Revenue', color='tab:orange')
ax2.tick_params(axis='y', labelcolor='tab:orange')

fig.tight_layout()
plt.legend(loc='upper left')
plt.show()

# Plotting Room Revenue and ADR
fig, ax1 = plt.subplots(figsize=(14, 7))

ax1.set_title('Monthly Room Revenue and ADR')
ax1.set_xlabel('Date')

ax1.plot(monthly_data.index, monthly_data['Room_Revenue'], color='tab:blue', label='Room Revenue')
ax1.set_ylabel('Room Revenue', color='tab:blue')
ax1.tick_params(axis='y', labelcolor='tab:blue')

ax2 = ax1.twinx()
ax2.plot(monthly_data.index, monthly_data['ADR'], color='tab:orange', label='ADR')
ax2.set_ylabel('ADR', color='tab:orange')
ax2.tick_params(axis='y', labelcolor='tab:orange')

fig.tight_layout()
plt.legend(loc='upper left')
plt.show()

# Plotting Total Revenue and Occupancy Rate
fig, ax1 = plt.subplots(figsize=(14, 7))

ax1.set_title('Monthly Total Revenue and Occupancy Rate')
ax1.set_xlabel('Date')

ax1.plot(monthly_data.index, monthly_data['Total_Revenue'], color='tab:blue', label='Total Revenue')
ax1.set_ylabel('Total Revenue', color='tab:blue')
ax1.tick_params(axis='y', labelcolor='tab:blue')

ax2 = ax1.twinx()
ax2.plot(monthly_data.index, monthly_data['Occupancy_Rate'], color='tab:orange', label='Occupancy Rate')
ax2.set_ylabel('Occupancy Rate', color='tab:orange')
ax2.tick_params(axis='y', labelcolor='tab:orange')

fig.tight_layout()
plt.legend(loc='upper left')
plt.show()
```

# Seasonal Impact Calculation

```python
seasonal_impact = data.groupby(['Season', 'Holiday'])['Revenue'].mean()
print(seasonal_impact)
```

```python
seasonal_impact_df = seasonal_impact.reset_index()
seasonal_impact_df.columns = ['Season', 'Holiday', 'Average_Revenue']

# Pivot the DataFrame for the heatmap
heatmap_data = seasonal_impact_df.pivot(index='Season', columns='Holiday', values='Average_Revenue')

# Plot the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data, annot=True, fmt=".1f", cmap='YlGnBu', cbar_kws={'label': 'Average Revenue'})
plt.title('Heatmap of Average Revenue by Season and Holiday')
plt.show()
```
